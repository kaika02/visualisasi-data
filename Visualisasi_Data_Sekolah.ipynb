```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("sekolah.csv")
data.head()
```

```python
data.info()
```

# Bar Chart dan Pie Chart Jumlah Sekolah berdasarkan Tingkat Sekolah 

```python
unique_tingkat_sekolah = data['tingkat_sekolah'].unique()
unique_tingkat_sekolah.sort()
unique_tingkat_sekolah
```

```python
jml_tingkat_sekolah = []
for i in unique_tingkat_sekolah:
  jml_tingkat_sekolah.append(data[data['tingkat_sekolah']==i].jumlah_sekolah.sum())
```

```python
plt.figure(figsize=(7,7))
plt.bar(unique_tingkat_sekolah, jml_tingkat_sekolah)
plt.title('Jumlah Sekolah berdasarkan Tingkat Sekolah')
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(jml_tingkat_sekolah, autopct='%1.1f%%')
plt.legend(unique_tingkat_sekolah)
plt.title('Jumlah Sekolah berdasarkan Tingkat Sekolah')
plt.show()
```
